```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster  # Importer MarkerCluster
from folium import DivIcon  # Importer DivIcon

# Crer une nouvelle colonne 'marker_color' base sur la colonne 'class'
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')

# Crer un objet MarkerCluster
marker_cluster = MarkerCluster().add_to(site_map)

# Ajouter des cercles et des marqueurs pour chaque site de lancement dans le DataFrame
for index, row in spacex_df.iterrows():
    coordinate = [row['Lat'], row['Long']]
    
    # Crer un cercle pour le site de lancement
    launch_circle = folium.Circle(
        coordinate,
        radius=1000,
        color=row['marker_color'],  # Utiliser la couleur dfinie
        fill=True,
        fill_opacity=0.5
    ).add_child(folium.Popup(f"{row['Launch Site']}<br>Status: {'Success' if row['class'] == 1 else 'Failure'}"))
    
    # Ajouter le cercle  la carte
    site_map.add_child(launch_circle)

    # Crer un marqueur pour le site de lancement
    launch_marker = folium.Marker(
        location=coordinate,
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12; color:{};"><b>{}</b></div>'.format(row['marker_color'], row['Launch Site']),
        )
    )
    
    # Ajouter le marqueur au MarkerCluster
    marker_cluster.add_child(launch_marker)

# Ajouter marker_cluster  la carte actuelle
site_map.add_child(marker_cluster)

# Enregistrer la carte dans un fichier HTML
site_map.save('launch_sites_map.html')

# Afficher la carte (si vous utilisez un notebook Jupyter, utilisez la ligne suivante)
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.


```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
from folium import DivIcon
from folium.plugins import MousePosition  # Importing MousePosition
from math import sin, cos, sqrt, atan2, radians

# Function to calculate the distance between two points
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371.0  # Radius of Earth in km
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Example coastline coordinate (the closest coastline point)
coastline_lat, coastline_lon = 28.56367, -80.57163  

# Initialize the map centered on the coastline
site_map = folium.Map(location=[coastline_lat, coastline_lon], zoom_start=10)

# Create a DataFrame for the specific launch site
data = {
    'Launch Site': ['CCAFS SLC-40'],
    'Lat': [28.563197],  # Launch site coordinates
    'Long': [-80.576820],
    'class': [1]  # Example success (1)
}
spacex_df = pd.DataFrame(data)

# Create a new column 'marker_color' based on the 'class' column
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')

# Initialize a MarkerCluster
marker_cluster = MarkerCluster().add_to(site_map)

# Add a marker for the coastline
coastline_marker = folium.Marker(
    location=[coastline_lat, coastline_lon],
    popup="Coastline",
    icon=folium.Icon(color='blue')  # You can change the color as desired
)
coastline_marker.add_to(site_map)  # Add coastline marker to the map

# Add circles and markers for each launch site in the DataFrame
for index, row in spacex_df.iterrows():
    coordinate = [row['Lat'], row['Long']]
    
    # Calculate distance from the launch site to the coastline
    distance = calculate_distance(row['Lat'], row['Long'], coastline_lat, coastline_lon)

    # Create a circle for the launch site
    launch_circle = folium.Circle(
        coordinate,
        radius=1000,
        color=row['marker_color'],  # Use the defined color
        fill=True,
        fill_opacity=0.5
    ).add_child(folium.Popup(f"{row['Launch Site']}<br>Status: {'Success' if row['class'] == 1 else 'Failure'}"))
    
    # Add the circle to the map
    site_map.add_child(launch_circle)

    # Create a marker for the launch site with distance info in the popup
    launch_marker = folium.Marker(
        location=coordinate,
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12; color:{};"><b>{}</b></div>'.format(row['marker_color'], row['Launch Site']),
        )
    )
    
    # Add the distance to the popup
    launch_marker.add_child(folium.Popup(f"Distance to Coastline: {distance:.2f} KM"))
    
    # Add the marker to the MarkerCluster
    marker_cluster.add_child(launch_marker)

    # Create and add a PolyLine from the launch site to the coastline
    coordinates = [coordinate, [coastline_lat, coastline_lon]]  # Coordinates for the PolyLine
    lines = folium.PolyLine(locations=coordinates, weight=2, color='purple')  # Change the color as desired
    site_map.add_child(lines)  # Add the PolyLine to the map

# Add MousePosition to the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)

# Add marker_cluster to the current map
site_map.add_child(marker_cluster)

# Save the map to an HTML file
site_map.save('launch_site_CCAFS_SLC_40_map_with_mouse_position.html')

# Display the map (if using a Jupyter notebook, use the line below)
site_map  # Uncomment this if you are running in a Jupyter notebook
```

```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster, MousePosition  # Importing MousePosition
from folium import DivIcon
from math import sin, cos, sqrt, atan2, radians

# Function to calculate the distance between two points
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371.0  # Radius of Earth in km
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Updated coastline coordinate
coastline_lat, coastline_lon = 28.56418, -80.56812  

# Railway and highway coordinates
railway_coordinates = (28.57306, -80.58525)
highway_coordinates = (28.56852, -80.57168)

# Initialize the map centered on the coastline
site_map = folium.Map(location=[coastline_lat, coastline_lon], zoom_start=10)

# Create a DataFrame for the specific launch site
data = {
    'Launch Site': ['CCAFS SLC-40'],
    'Lat': [28.563197],  # Launch site coordinates
    'Long': [-80.576820],
    'class': [1]  # Example success (1)
}
spacex_df = pd.DataFrame(data)

# Create a new column 'marker_color' based on the 'class' column
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')

# Initialize a MarkerCluster
marker_cluster = MarkerCluster().add_to(site_map)

# Add a marker for the coastline
coastline_marker = folium.Marker(
    location=[coastline_lat, coastline_lon],
    popup="Coastline",
    icon=folium.Icon(color='blue')  # You can change the color as desired
)
coastline_marker.add_to(site_map)  # Add coastline marker to the map

# Add markers for railway and highway
railway_marker = folium.Marker(
    location=railway_coordinates,
    popup="Railway",
    icon=folium.Icon(color='orange')  # Orange color for railway
)
railway_marker.add_to(site_map)  # Add railway marker to the map

highway_marker = folium.Marker(
    location=highway_coordinates,
    popup="Highway",
    icon=folium.Icon(color='red')  # Red color for highway
)
highway_marker.add_to(site_map)  # Add highway marker to the map

# Add circles and markers for each launch site in the DataFrame
for index, row in spacex_df.iterrows():
    coordinate = [row['Lat'], row['Long']]
    
    # Calculate distances from the launch site to the coastline, railway, and highway
    distance_coastline = calculate_distance(row['Lat'], row['Long'], coastline_lat, coastline_lon)
    distance_railway = calculate_distance(row['Lat'], row['Long'], *railway_coordinates)
    distance_highway = calculate_distance(row['Lat'], row['Long'], *highway_coordinates)

    # Create a circle for the launch site
    launch_circle = folium.Circle(
        coordinate,
        radius=1000,
        color=row['marker_color'],  # Use the defined color
        fill=True,
        fill_opacity=0.5
    ).add_child(folium.Popup(f"{row['Launch Site']}<br>Status: {'Success' if row['class'] == 1 else 'Failure'}"))
    
    # Add the circle to the map
    site_map.add_child(launch_circle)

    # Create a marker for the launch site with distance info in the popup
    launch_marker = folium.Marker(
        location=coordinate,
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12; color:{};"><b>{}</b></div>'.format(row['marker_color'], row['Launch Site']),
        )
    )
    
    # Add distances to the popup
    launch_marker.add_child(folium.Popup(f"""
        Distance to Coastline: {distance_coastline:.2f} KM<br>
        Distance to Railway: {distance_railway:.2f} KM<br>
        Distance to Highway: {distance_highway:.2f} KM
    """))
    
    # Add the marker to the MarkerCluster
    marker_cluster.add_child(launch_marker)

    # Create and add a PolyLine from the launch site to the coastline
    coordinates = [coordinate, [coastline_lat, coastline_lon]]  # Coordinates for the PolyLine
    lines = folium.PolyLine(locations=coordinates, weight=2, color='purple')  # Change the color as desired
    site_map.add_child(lines)  # Add the PolyLine to the map

    # Create and add a PolyLine from the launch site to the railway
    railway_line = folium.PolyLine(locations=[coordinate, railway_coordinates], weight=2, color='orange')
    site_map.add_child(railway_line)

    # Create and add a PolyLine from the launch site to the highway
    highway_line = folium.PolyLine(locations=[coordinate, highway_coordinates], weight=2, color='red')
    site_map.add_child(highway_line)

# Add MousePosition to the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)

# Add marker_cluster to the current map
site_map.add_child(marker_cluster)

# Save the map to an HTML file
site_map.save('launch_site_CCAFS_SLC_40_map_with_mouse_position.html')

# Display the map (if using a Jupyter notebook, use the line below)
site_map  # Uncomment this if you are running in a Jupyter notebook
```
